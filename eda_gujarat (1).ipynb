### using wordcloud to display most common/frequent words in Title

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words='english')   # remove common words
bow_matrix = vectorizer.fit_transform(df['Title'].astype(str))
```

```python
word_counts = bow_matrix.sum(axis=0).A1
words = vectorizer.get_feature_names_out()
```

```python
freq_df = pd.DataFrame({'word': words, 'count': word_counts})
freq_df = freq_df.sort_values(by='count', ascending=False)
```

```python
from wordcloud import WordCloud
```

```python
word_freq_dict = dict(zip(freq_df['word'], freq_df['count']))

wc = WordCloud(width=800, height=400, background_color='white')
wc.generate_from_frequencies(word_freq_dict)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```
